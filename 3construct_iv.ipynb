Create two functions to convert between week number and calendar year and month

```python
#this is the file to switch between week number in NLSY79 and calendar year, month, and day
xl_file = pd.ExcelFile('input_data/continuous_week_crosswalk_2012.xlsx') 

weekdf = xl_file.parse('weekdates1')
weekdf.head()
```

```python
def get_week_num(year, month, which, df=weekdf):
    """
    Given a calendar year and month, return the corresponding week number in NLSY79.
    
    If which=='first', return the first week in that month.
    If which=='last', return the last week in that month. 
    """
    
    the_month = weekdf[(weekdf['Week Start:\nMonth']==month)&(weekdf['Week Start: \nYear']==year)]
    
    if which == 'first':
        return int(the_month.iloc[0]['Continuous \nWeek Number'])
    elif which == 'last':
        return int(the_month.iloc[len(the_month)-1]['Continuous \nWeek Number'])
```

```python
def get_year_month(weeknum, df=weekdf):
    """
    Given a week number, return the corresponding calendar month and year (yyyy, m).
    """
    
    the_week = weekdf[weekdf['Continuous \nWeek Number']==weeknum]
    return int(the_week.iloc[0]['Week Start: \nYear']), int(the_week.iloc[0]['Week Start:\nMonth'])
```

```python
def find_rate(rate_row, year, j):
    """
    Given a row from job rate dataframe, year, and job number, return the job pay rate. 
    If the rate is negative/missing, return None. 
    """
    
    cps_id_row = cps_id_df[cps_id_df.caseid_1979==rate_row['caseid_1979']].iloc[0]
    try:
        cps_id = cps_id_row['qes%s_52_%s'%(j, year)]
    except KeyError: #no cps job indicators
        r = rate_row['hrp%s_%s'%(j, year)]
    else:
        if cps_id==1: #the jth job is equivalent to cps job
            r_cps = rate_row['cpshrp_%s'%year]
            r = rate_row['hrp%s_%s'%(j, year)]
            
            if r>=0 and r_cps>=0 and r!=r_cps:
                print "cps job and jth job don't pay the same", rate_row['caseid_1979'], year, j 
                #this never happened. never got printed
            elif r<0:
                if r_cps<0: #both jth job and cps job have missing pay
                    r = None
                else: #if jth job is invalid, but cps job is valid, use cps job rate as the jth job rate
                    r = r_cps
        else: #jth job is not cps job
            r = rate_row['hrp%s_%s'%(j, year)]
    
    if r<0:
        return None
    return r
```

```python
#version 2 - final

def get_job_info(i):
    """
    Given respondent id, find the latest valid job that the respondent took one year prior to first birth, 
    and return hourly rate, job tenure, and how many weeks are between the job I found and one year before the first birth.
    """
    
    dob_row = analytic_df[analytic_df.caseid_1979==i].iloc[0]
    dob_m = int(dob_row['c1dob_m_xrnd'])
    dob_y = int(dob_row['c1dob_y_xrnd'])
    
    #find the week number that is one year pre-birth
    #assume birth took place in the last week of the month
    week = get_week_num(dob_y-1, dob_m, 'last') 
    
    jobs = []
    job_row = jobdates_df[jobdates_df.caseid_1979==i].iloc[0]
    
    for year in years:   
        for j in range(1, 6):
            start = int(job_row['start_wk_%i_job%02d_xrnd'%(year, j)])
            #end = int(job_row['stop_wk_%i_job%02d_xrnd'%(year, j)])
            # some jobs have missing stop dates, but it's irrelevant to what I'm doing here
            
            if start >= 0 and start <= week: #the job has a valid starting week number and the starting week number is one year pre-birth
                jobs.append((start, j, year))
    
    #sort jobs. the first is the most recent/latest job
    jobs = sorted(jobs, key=itemgetter(0), reverse=True) 
    
    
    if len(jobs)==0: # no job found one year prior to birth
        # return (rate, job_year, tenure, working, gap)
        return (0, None, 0, 0, 0)

    rate_row = hrrate_df[hrrate_df.caseid_1979==i].iloc[0]
    rate = find_rate(rate_row, jobs[0][2], jobs[0][1]) #hourly rate of the job closest to one year pre-birth
            
    start = int(job_row['start_wk_%i_job%02d_xrnd'%(jobs[0][2], jobs[0][1])])
    end = int(job_row['stop_wk_%i_job%02d_xrnd'%(jobs[0][2], jobs[0][1])])
    
    gap = week-start # number of weeks between one year prior to birth and start week of job
    
    tenure = None
    job_year = None
    working = None #was repsondent working one year pre-birth?
    if end < 0 or end < start: #missing values or incorrect starting/end date
        #assume: 
        job_year = dob_y-1 
        tenure = week - start + 1
        working = 1
    elif end < week: #job ended before one year prior to birth
        tenure = end - start + 1
        job_year = get_year_month(end)[0]
    else: #job ended after one year prior to birth
        tenure = week - start + 1
        job_year = dob_y-1

    #working status one year pre-birth
    if start>=0 and end>=0 and end>=start: #valid start and end dates of the most recent job
        latest_job_span = range(start, end+1)
        # one year prior to birth time
        prebirth_span = range(get_week_num(dob_y-1, dob_m, 'first'), 
                              get_week_num(dob_y-1, dob_m, 'last')+1) 
        if len(set(latest_job_span)&set(prebirth_span))>0:
            working = 1
        else:
            working = 0
    else: 
        assert working==1 #we assumed the five invalid cases to be working
        
        
    return (rate, job_year, tenure, working, gap)
```

```python
#helper functions
def is_missing(v):
    """Return whether v is a missing value (-1, -2, or -3)."""
    return v in [-1, -2, -3]
                 
def after(y1, m1, y2, m2):
    """Given two dates (year-month), return True if the second date is after the first date. """
    
    if is_missing(y2):
        return None
    else:
        if y2>y1:
            return True
        elif y2<y1:
            return False
        elif y2==y1:
            if is_missing(m2):
                return None
            else:
                return m2>m1
```

```python
# marital status
# 0 - never married before, 1 - married, 2 - divorced/separated/widowed

def get_marstat_prev(i):
    """
    This function returns the marital status 12 months prior to first birth, 
    determined using marriage dates.
    """
    
    dob_row = analytic_df[analytic_df.caseid_1979==i].iloc[0]
    mar_row = marstat_df[marstat_df.caseid_1979==i].iloc[0]
    m = int(dob_row['c1dob_m_xrnd'])
    y = int(dob_row['c1dob_y_xrnd'])-1
    
    #Documentation:
        # -999: NEVER MARRIED
        # -998: 1ST MARRIAGE NEVER ENDED
        # -997: 1ST MARRIAGE ENDED NO 2ND MARRIAGE
        # -996: 2ND MARRIAGE NEVER ENDED
        # -995: 2ND MARRIAGE ENDED NO 3RD MARRIAGE
        # 0 TO 1979: <=1979
        # 1980, etc.
    
    ####################################
    ### Consider various situations: ###
    ####################################
    
    #never married
    if mar_row['yrbg1m_xrnd'] == -999: 
        return 0
    
    # first marriage never ended
    if mar_row['yren1m_xrnd'] == -998: 
        comp = after(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'])
        if comp==True: #first marriage started after the given year
            return 0 
        elif comp==None: 
            return None
        else:
            return 1
    
    # 1ST MARRIAGE ENDED NO 2ND MARRIAGE
    if mar_row['yrbg2m_xrnd'] == -997: 
        comp = after(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'])
        if comp==True:
            return 0
        elif comp==None:
            return None
        else:
            comp2 = after(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd']) #second comparison
            if comp2==True: #first marriage ended after the given time
                return 1
            elif comp2==None:
                return None
            else: #first marriage ended before or on the given time
                return 2 
    
    #2ND MARRIAGE NEVER ENDED
    if mar_row['yren2m_xrnd'] == -996:
        comp = after(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'])
        if comp==True:
            return 0
        elif comp==None:
            return None
        else:
            comp2 = after(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'])
            if comp2==True: #first marriage ended after the given time
                return 1
            elif comp2==None:
                return None
            else: 
                comp3 = after(y, m, mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd'])
                if comp3==True:
                    return 2
                elif comp3==None:
                    return None
                else:
                    return 1

    #2ND MARRIAGE ENDED NO 3RD MARRIAGE                
    if mar_row['yrbg3m_xrnd'] == -995:
        comp = after(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'])
        if comp==True:
            return 0
        elif comp==None:
            return None
        else:
            comp2 = after(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'])
            if comp2==True: #first marriage ended after the given time
                return 1
            elif comp2==None:
                return None
            else: 
                comp3 = after(y, m, mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd'])
                if comp3==True:
                    return 2
                elif comp3==None:
                    return None
                else:
                    comp4 = after(y, m, mar_row['yren2m_xrnd'], mar_row['moen2m_xrnd'])
                    if comp4==True:
                        return 1
                    elif comp4==None:
                        return None
                    else:
                        return 2
                    

    #THERE IS A THIRD MARRIAGE        
    if mar_row['yrbg3m_xrnd']>=0:
        comp = after(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'])
        if comp==True:
            return 0
        elif comp==None:
            return None
        else:
            comp2 = after(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'])
            if comp2==True: #first marriage ended after the given time
                return 1
            elif comp2==None:
                return None
            else: 
                comp3 = after(y, m, mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd'])
                if comp3==True:
                    return 2
                elif comp3==None:
                    return None
                else:
                    comp4 = after(y, m, mar_row['yren2m_xrnd'], mar_row['moen2m_xrnd'])
                    if comp4==True:
                        return 1
                    elif comp4==None:
                        return None
                    else:
                        comp5 = after(y, m, mar_row['yrbg3m_xrnd'], mar_row['mobg3m_xrnd'])
                        if comp5==True:
                            return 2
                        elif comp5==None:
                            return None
                        else:
                            return 'unknown 3rd marriage'
                        
    return 'uncaptured'
```

##Average time between first and second birth

Assume that the child was born at the end of the month (i.e. if a woman reports that her baby was born in March, we call April her first month of parenthood, and we start the employment history with the week that includes April 1). For subsequent births, also assume that the birth occurred at the end of the month in which it was reported.

```python
def find_month(start_y, start_m, y, m):
    """
    Return the number of months between (start_y, start_m) and (y, m).
    (start_y, start_m) is earlier than (y, m).
    """
    
    i = 0
    while (y!=start_y or m!=start_m):
        if y > start_y:
            if start_m < 12:
                start_m += 1
            elif start_m == 12:
                start_m = 1
                start_y += 1
        elif y == start_y:
            if m > start_m:
                start_m += 1
        i += 1
    
    return i
            
#for example,
find_month(1989, 2, 1990, 4)
```

```python
def is_valid_m(m):
    """Return whether m is a valid month."""
    return (m>=1 and m<=12)
    
    
def is_valid_y(y):
    """Return whether y is a valid year."""
    return (y>1970 and y<=2013)


def find_marhist(i):
    """
    Given an id, find the type of woman's marital history.
    For now, just check whether the marriage dates are complete and valid.
    """
    
    row = sub_marriage_df[sub_marriage_df.caseid_1979==i].iloc[0]
    vals = row.values.tolist()[1:]
    
    ######## assume missing month to be July ########
    for j in range(0, 10, 2):
        if vals[j] in [-1, -2, -3, -4, -5]:
            vals[j] = 7
    
    # type 1: never married, ever
    if len(set(vals))==1 and -999 in vals:
        return 1
    
    # type 2: married once and never divorced
    if is_valid_m(vals[0]) and is_valid_y(vals[1]):
        if len(set(vals[2:]))==1 and -998 in vals:
            return 2
    
    # type 3: married once and divorced, no second marriage
    if is_valid_m(vals[0]) and is_valid_y(vals[1]) and \
        is_valid_m(vals[2]) and is_valid_y(vals[3]):
            if len(set(vals[4:]))==1 and -997 in vals:
                return 3
    
    # type 4: second marriage never ended
    if is_valid_m(vals[0]) and is_valid_y(vals[1]) and \
        is_valid_m(vals[2]) and is_valid_y(vals[3]) and \
        is_valid_m(vals[4]) and is_valid_y(vals[5]):
            if len(set(vals[6:]))==1 and -996 in vals:
                return 4
    
    # type 5: second marriage ended, no third marriage
    if is_valid_m(vals[0]) and is_valid_y(vals[1]) and \
        is_valid_m(vals[2]) and is_valid_y(vals[3]) and \
        is_valid_m(vals[4]) and is_valid_y(vals[5]) and \
        is_valid_m(vals[6]) and is_valid_y(vals[7]):
        if len(set(vals[8:]))==1 and -995 in vals:
            return 5
    
    # type 6: third marriage began
    if is_valid_m(vals[0]) and is_valid_y(vals[1]) and \
        is_valid_m(vals[2]) and is_valid_y(vals[3]) and \
        is_valid_m(vals[4]) and is_valid_y(vals[5]) and \
        is_valid_m(vals[6]) and is_valid_y(vals[7]) and \
        is_valid_m(vals[8]) and is_valid_y(vals[9]):
        return 6
    
    # type 0: there are invalid marital dates
    return 0
```

```python
#some helper functions to determine order of two dates

def before(y, m, base_y, base_m):
    """Return True if y, m is before the base y, m (exclusively)"""

    given = datetime(y, m, 1)
    base = datetime(base_y, base_m, 1)
    return given<base


def between(y, m, start_y, start_m, end_y, end_m):
    """Return True if y, m is between the start y, m and end y, m (inclusively)"""
    
    given = datetime(y, m, 1)
    start = datetime(start_y, start_m, 1)
    end = datetime(end_y, end_m, 1)
    
    return (given>=start and given<=end)


def after(y, m, base_y, base_m):
    """Return True if y, m is after the base y, m (exclusively)"""
    
    given = datetime(y, m, 1)
    base = datetime(base_y, base_m, 1)
    return given>base
```

```python
data = defaultdict(list) # dataframe for monthly marital status

for i in sample:
    marhist = find_marhist(i)
    row = monthly_df[monthly_df.id==i].iloc[0]
    mar_row = sub_marriage_df[sub_marriage_df.caseid_1979==i].iloc[0]
    
    ######## assume and change missing month to be July ########
    for j in range(1, 10, 2):
        if mar_row[j] in [-1, -2, -3, -4, -5]:
            mar_row[j] = 7
    
    #================================================================================
    if marhist==0: #has missing marriage dates. missing month is already assumed to be July 
        continue
    
    #================================================================================
    elif marhist==1: #always single
        data['id'].append(i)
        
        for j in range(216):
            data['month%s'%j].append(0)
    
    #================================================================================
    elif marhist==2: #married once and never divorced 
        data['id'].append(i)
        
        y = row['start_y']
        m = row['start_m']
        for j in range(216):
            if before(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd']):
                data['month%s'%j].append(0) #single, before getting married
            else:
                data['month%s'%j].append(1) #married
            
            if m == 12:
                m = 1
                y += 1
            else:
                m += 1
                
    #================================================================================
    # type 3: married once and divorced, no second marriage
    elif marhist==3: 
        data['id'].append(i)
        
        y = row['start_y']
        m = row['start_m']
        for j in range(216):
            
            if before(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd']):
                data['month%s'%j].append(0)#single, before getting married
            elif after(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd']):
                data['month%s'%j].append(2) #divorced/widowed
            else:
                data['month%s'%j].append(1) #married
            
            if m == 12:
                m = 1
                y += 1
            else:
                m += 1
    
    #================================================================================
    # type 4: second marriage never ended
    elif marhist==4:
        data['id'].append(i)
        
        y = row['start_y']
        m = row['start_m']
        for j in range(216):
            if before(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd']):
                data['month%s'%j].append(0)
            elif between(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'], mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd']):
                data['month%s'%j].append(1)
            elif between(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'], mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd']):
                data['month%s'%j].append(2)
            else:
                data['month%s'%j].append(1) 
                
            if m == 12:
                m = 1
                y += 1
            else:
                m += 1
                #y remains the same 
                
    
    #================================================================================
    # type 5: second marriage ended, no third marriage
    elif marhist==5:
        data['id'].append(i)
        
        y = row['start_y']
        m = row['start_m']
        for j in range(216):
            if before(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd']):
                data['month%s'%j].append(0) #single, before getting married
            elif between(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'], mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd']):
                data['month%s'%j].append(1)
            elif between(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'], mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd']):
                data['month%s'%j].append(2) 
            elif between(y, m, mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd'], mar_row['yren2m_xrnd'], mar_row['moen2m_xrnd']):
                data['month%s'%j].append(1)
            else:
                data['month%s'%j].append(2) 
                
            if m == 12:
                m = 1
                y += 1
            else:
                m += 1
                #y remains the same 
    
    #================================================================================
    # type 6: third marriage began. we don't know what happened to third marriage end time.    
    elif marhist==6:
        row_yearly = sub_df[sub_df.caseid_1979==i].iloc[0]
        vals_after = [row_yearly['marstat_%s'%y] for y in years if y>mar_row['yrbg3m_xrnd']]
        missing = [v<0 for v in vals_after]
        if sum(missing) > 0: #there is missing value in the yearly statuses after third marriage
            continue
            
        else:
            data['id'].append(i)

            y = row['start_y']
            m = row['start_m']
            
            married = True
            
            for j in range(216):
                if before(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd']):
                    data['month%s'%j].append(0) #single, before getting married
                elif between(y, m, mar_row['yrbg1m_xrnd'], mar_row['mobg1m_xrnd'], mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd']):
                    data['month%s'%j].append(1)
                elif between(y, m, mar_row['yren1m_xrnd'], mar_row['moen1m_xrnd'], mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd']):
                    data['month%s'%j].append(2) 
                elif between(y, m, mar_row['yrbg2m_xrnd'], mar_row['mobg2m_xrnd'], mar_row['yren2m_xrnd'], mar_row['moen2m_xrnd']):
                    data['month%s'%j].append(1)
                elif between(y, m, mar_row['yren2m_xrnd'], mar_row['moen2m_xrnd'], mar_row['yrbg3m_xrnd'], mar_row['mobg3m_xrnd']):
                    data['month%s'%j].append(2) 
                else: 
                    
                    vals_after_dict = {y:row_yearly['marstat_%s'%y] for y in years if y>mar_row['yrbg3m_xrnd']}

                    #treat separated as divorced                            
                    if married: #previous status is married. check if status needs to be flipped to divorced
                        if y in vals_after_dict.keys():
                            if vals_after_dict[y] in [2, 3]: #divorced. assume divorce happened in July
                                if m>=7:
                                    married = False
                    else: #previous status is divorced. check if status needs to be flipped to married
                        if y in vals_after_dict.keys():
                            if vals_after_dict[y]==1: #married. assume marriage happened in July
                                if m>=7:
                                    married = True
                            
                            
                    if married:
                        data['month%s'%j].append(1)
                    else:
                        data['month%s'%j].append(2)

                if m == 12:
                    m = 1
                    y += 1
                else:
                    m += 1
                    #y remains the same 
        
```
