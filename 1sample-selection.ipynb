```python
#setup
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
from collections import defaultdict
```

###Select sample

```python
dob_df = pd.read_csv('childbirth.csv')
print dob_df.shape
dob_df.head()
```

```python
#drop males
#1-male; 2-female

sample_df = dob_df[dob_df.sample_sex_1979==2]
print sample_df.shape
```

```python
#drop subsamples that weren’t followed over the entire survey period
print sample_df.sample_id_1979.unique()
```

```python
sample_df = sample_df[(~sample_df.sample_id_1979.isin([9, 12, 15, 16, 17, 18, 19, 20]))]
print sample_df.sample_id_1979.unique()
print sample_df.shape
```

```python
#distribution of child birth dates
gby = sample_df.groupby('c1dob_y_xrnd')
gby.count()
```

```python
#number of women who had first birth too early - before 1980 (EXCLUSIVE)
sample_df[(sample_df.c1dob_y_xrnd>0)&(sample_df.c1dob_y_xrnd<1980)].shape
```

```python
#number of women who had first birth too late - after 1994 (EXCLUSIVE)
sample_df[(sample_df.c1dob_y_xrnd>0)&(sample_df.c1dob_y_xrnd>1994)].shape
```

###Determine last valid year from reasons for non-interview variables 

```python
rni_df = pd.read_csv('reasonnonint.csv')
rni_df.head()
```

```python
#put respondent ids in a list 
sample = sample_df.caseid_1979.values.tolist()
```

```python
#determine last valid year
last_yr = []

for i in sample: #iterate through the sample

    #find the row in reason-noninterview dataframe that corresponds to the given id
    row = rni_df[rni_df.caseid_1979==i].iloc[0]
    
    j = -1
    while row[j] != -4:
        if row.index[j-1] == 'caseid_1979': 
            break
        else:
            j -= 1

    
    col_name = row.index[j]
    last_yr.append(int(col_name[-4:]))
        
print set(last_yr)
```

```python
sample_df['last_yr'] = last_yr
sample_df.head()
```

```python
#distribution of last valid year
gby = sample_df.groupby('last_yr')
gby.count()
```

###More specifics on women who never had childbirth in our observation period

```python
#women who never had children
childless_df = sample_df[sample_df.c1dob_y_xrnd==-4]
```

```python
age_df = pd.read_csv('ageatint.csv')
age_df.head()
```

```python
childless_df = childless_df.merge(age_df[['caseid_1979', 'ageatint_1979']], on='caseid_1979', how='left')
childless_df.head()
```

```python
childless_df['year_age35'] = childless_df.ageatint_1979.map(lambda x: 35-x+1979)
childless_df.head()
```

```python
def attr_before35(row):
    """
    Return True if respondent left study before turning age 35.
    Return False otherwise. 
    """
    return row['last_yr']<row['year_age35']

childless_df['attr_before35'] = childless_df.apply(attr_before35, axis=1)
```

```python
#among childless women, distribution of "being truly childless" 
#i.e. remain in the study until 35 and remain childless
gby_attr35 = childless_df.groupby('attr_before35')
gby_attr35.count()
```

###Select analytic sample

```python
analytic_df = sample_df[(sample_df.c1dob_y_xrnd>=1980)&(sample_df.c1dob_y_xrnd<=1994)]
print analytic_df.shape
```

```python
analytic_df['year_child18'] = analytic_df.c1dob_y_xrnd.map(lambda x: x+18)
analytic_df.head()
```

```python
def attr_before_child18(row):
    """
    Return True if respondent left study before her first child turned age 18.
    Return False otherwise. 
    """
    return row['last_yr']<row['year_child18']

analytic_df['attr_child18'] = analytic_df.apply(attr_before_child18, axis=1)
analytic_df.head()
```

```python
#distribution of attrition before first child turning age 18
gby_attr_child18 = analytic_df.groupby('attr_child18')
gby_attr_child18.count()
```

```python
analytic_df = analytic_df[analytic_df.attr_child18==False]
print analytic_df.shape
```

The final sample consists of 2459 mothers.

```python
analytic_df.to_csv('analytic_df.csv', index=False)
```
