```python
def get_week_num(year, month, which, df=weekdf):
    """Given a year and month in the calendar, 
    return the corresponding week number in NLSY79
    using coutinuous_week_crosswalk.xlsx
    
    if which=='first', return the first week in that month
    if which=='last', return the last week in that month 
    """

    
    #column names in weekdf:
    #u'Week Start:\nMonth'
    #u'Week Start: \nDay'
    #u'Week Start: \nYear'
    #u'Calendar Year \nWeek Number '
    #u'Continuous \nWeek Number'
    
    that_month = weekdf[(weekdf['Week Start:\nMonth']==month)&(weekdf['Week Start: \nYear']==year)]
    
    if which == 'first':
        return int(that_month.iloc[0]['Continuous \nWeek Number'])
    elif which == 'last':
        return int(that_month.iloc[len(that_month)-1]['Continuous \nWeek Number'])
```

```python
def get_year_month(weeknum, df=weekdf):
    """Given a week number, return the corresponding month and year (yyyy, m).
    """
    
    #column names in weekdf:
    #u'Week Start:\nMonth'
    #u'Week Start: \nDay'
    #u'Week Start: \nYear'
    #u'Calendar Year \nWeek Number '
    #u'Continuous \nWeek Number'
    
    that_week = weekdf[weekdf['Continuous \nWeek Number']==weeknum]
    return int(that_week.iloc[0]['Week Start: \nYear']), int(that_week.iloc[0]['Week Start:\nMonth'])
```

```python
def list_duplicates_of(seq,item):
    """
    a function to return the indexes of duplicate items in a list (as a list)
    """
    start_at = -1
    locs = []
    while True:
        try:
            loc = seq.index(item,start_at+1)
        except ValueError:
            break
        else:
            locs.append(loc)
            start_at = loc
    return locs
```

```python
def check_distribution(value, df, start_value):
    """
    A function to exmine the distribution of the given value in the given dataframe
    
    start_value is the column number starting from which are our actual data. 
    e.g. here we want only to examine the distribution of certain values in the weekly status,
    so the columns id, start_y, start_m, and start_week are not relevant, and we want to skip
    those columns, so the start_value will be 4.
    """   
    
    ###how many respondents have the given value in their sequences?
    indices_by_person = []
    for i, row in df.iterrows():
        
        if value in row.values[start_value:]: 
            seq = row.values[start_value:].tolist()
            
            #find the list of the given value's positions in the sequence
            indices_by_person.append(list_duplicates_of(seq, value))  
            
    print '%s respondents have some weeks of status %s'%(len(indices_by_person), value)
    print
    
    ###how many such weeks are in each respondent's sequence?
    count_by_person = [len(l) for l in indices_by_person]
    plt.hist(count_by_person, bins=50)
    plt.title('Count of Weeks of Status %s per Respondent\n(Not Considering Whether Continuously distributed or Not)'%value)
    plt.ylabel('Frequency')
    remove_border()
    plt.show()    
    print 'The number of weeks with %s status per respondent'%value
    print 'Min:', min(count_by_person)
    print 'Max:', max(count_by_person)
    print 'Mean:', np.mean(count_by_person)
    print 'Quantiles:', np.percentile(count_by_person, 25), \
                        np.percentile(count_by_person, 50), \
                        np.percentile(count_by_person, 75)
    print
 
    ###Overall gap lengths of the given status?
    gaps_by_person = [] 
    total_gaps = []
    for indices in indices_by_person:
        gap_len_for_this_person = []
        start = 0 #count the start pos of each continuous gap
        while start < len(indices):
            i = 0
            while start+i+1<len(indices) and indices[start+i+1]==indices[start+i]+1:
                i += 1
            gap_len_for_this_person.append(i+1)
            total_gaps.append(i+1)
            start += i+1
        gaps_by_person.append(gap_len_for_this_person)    
    
    plt.hist(total_gaps, bins=50)
    plt.title('Lengths of Spells of Status %s'%value)
    plt.ylabel('Frequency')
    remove_border()   
    plt.show()
    print 'The total number of spells of status %s: %s'%(value, len(total_gaps))        
    print 'The lengths of spells of status %s'%value
    print 'Min:', min(total_gaps)
    print 'Max:', max(total_gaps)
    print 'Mean:', np.mean(total_gaps)
    print 'Quantiles:', np.percentile(total_gaps, 25), \
                    np.percentile(total_gaps, 50), \
                    np.percentile(total_gaps, 75)
    print

    
    ###how many 'gaps' of the given status per respondent?
    count_gaps_by_person = [len(gaps_per_person) for gaps_per_person in gaps_by_person]
    plt.hist(count_gaps_by_person, bins=50)
    plt.title('Number of Spells of Status %s Per Respondents'%value)
    plt.ylabel('Frequency')
    remove_border()    
    plt.show()    
    print 'The number of spells of status %s per respondent'%value
    print 'Min:', min(count_gaps_by_person)
    print 'Max:', max(count_gaps_by_person)
    print 'Mean:', np.mean(count_gaps_by_person)
    print 'Quantiles:', np.percentile(count_gaps_by_person, 25), \
                    np.percentile(count_gaps_by_person, 50), \
                    np.percentile(count_gaps_by_person, 75)    
```

```python
def find_header(headers, h):
    """
    Given a string h, return the indices of all headers that match to h.
    """
     
    matched = [item for item in headers if item.startswith(h)]
    return matched
```

```python
def is_missing(value):
    """
    Return true if value is in the collection of missing values.
    """
     
    missings = [-1, -2, -3, -4, -5]
    return value in missings
```

```python
def convert_year(y):
    """
    Given a two-digit year, convert it back to 4-digit integer
    """
    
    if int(y) > 1900:
        return int(y)
    else:
        if 59 <= int(y) <= 99:
            return 1900+int(y)
        else:
            return 2000+int(y)
```

```python
def has_positive(l):
    """
    Return True if a positive number is in the list l
    """
    
    for n in l:
        if n > 0:
            return True
    return False
```

###2.4 If enrolled in school since last interview, how many months were the respondents in school?

Be careful: The codebook tells us that the set of questions "MONTHS ENROLLED IN SCHOOL SINCE LAST INT - given month and year" used two sets of answers --- 1-24 and 0-1.

```python
years = ['80','81','82','83','84','85','86','87','88','89','90','91',
         '92','93','94','96','98','00','02','04','06','08','10'] # No year 1979

counts = defaultdict(int)
how_many_month = defaultdict(int)

for i, row in school_df[school_df.id.isin(sample)].iterrows():
    
    if i%100 == 0:
        print i
    
    for yr in years:
        
        if row['inschoolsincelastint'+yr] == 1:
            
            res = 0
            
            for m in months:    
                col = 'enrollsincelastint_'+m+yr
                
                # Find columns containing the given month and year
                cols = find_header(school_df.columns, col) 
         
                data = [int(row[col]) for col in cols]

                if len(data) == 1:
                    if is_missing(data[0]):
                        counts['single, missing'] += 1
                    elif data[0] == 0:
                        counts['single, not in school'] += 1
                    elif data[0] > 0:
                        counts['single, in school'] += 1
                        res += 1
                    else: 
                        raise Exception, 'unexpected single status'
                        
                elif len(data) == 2:
                    if is_missing(data[0]) and is_missing(data[1]):
                        if data[0] == data[1]:
                            counts['two, same missing'] += 1
                        else:
                            counts['two, different missing'] += 1
                    
                    elif is_missing(data[0]) and not is_missing(data[1]):
                        if data[1] == 0:
                            counts['two, not in school and a missing'] += 1
                        else:
                            res += 1
                            counts['two, in school and a missing'] += 1
                            
                    elif not is_missing(data[0]) and is_missing(data[1]):
                        if data[0] == 0:
                            counts['two, not in school and a missing'] += 1
                        else:
                            res += 1
                            counts['two, in school and a missing'] += 1

                    elif not is_missing(data[0]) and not is_missing(data[1]):
                        if data[0] == 0 and data[1] == 0:
                            counts['two, same, not in school'] += 1
                        elif data[0] == 0 and data[1] > 0:
                            counts['two, one in school and one not in school'] += 1
                            res += 1
                        elif data[0] > 0 and data[1] == 0:
                            counts['two, one in school and one not in school'] += 1
                            res += 1
                        elif data[0] > 0 and data[1] > 0:
                            if data[0] == data[1]: # Unlikely
                                counts['two, same, in school'] += 1
                                res += 1
                            elif abs(int(data[0])-int(data[1])) == 12:
                                counts['two, same, in school'] += 1
                                res += 1
                            elif data[0] != data[1]:
                                counts['two, strange case'] += 1
                                res += 1
              
            how_many_month[res] += 1
```

###2.6 If a respondent had -5 for the question "enrolled in school since last interview" in a given year, what answer did she give in the preceding and following years?

```python
def get_status(row, yr):
    """Given a row of data and a year string (2-digit), 
    if yr == '79', return the response to the question 'currently in school.'
    Otherwise, return the response to the question 'enrolled since last interview.'
    """
    
    if yr == '79':
        return row['inschool79']
    else:
        return row['inschoolsincelastint'+yr]
```

```python
years = ['79','80','81','82','83','84','85','86','87','88','89','90','91',
         '92','93','94','96','98','00','02','04','06','08','10']

counts = defaultdict(int)

for i, row in school_df[school_df.id.isin(sample)].iterrows():
    
    for yr in years:
        
        if get_status(row,yr) == -5:
        
            if years.index(yr) == 0:
                next_yr = years[years.index(yr)+1]
                counts[('na',-5,get_status(row,next_yr))] += 1
            elif years.index(yr) == len(years)-1:
                prev_yr = years[years.index(yr)-1]
                counts[(get_status(row,prev_yr),-5,'na')] += 1
            else: # In the middle
                prev_yr = years[years.index(yr)-1]
                next_yr = years[years.index(yr)+1]
                counts[(get_status(row,prev_yr),-5,get_status(row,next_yr))] += 1
```

```python
# Version 1

def is_in_school(person_id, y, m):
    """
    Given a person id, year, and month, return True if the respondent 
    with the given id was in school during the given month.
    
    Return False, otherwise.
    """
    
    years = ['79','80','81','82','83','84','85','86','87','88','89','90','91',
             '92','93','94','96','98','00','02','04','06','08','10']
        
    # Find the rows needed
    school_row = school_df[school_df.id==person_id].iloc[0]
    age_row = age_df[age_df.id==person_id].iloc[0]
    
    # Convert the integer year into 2-digit string as in the years variable above
    str_y0 = str(y)[2:]
    str_y = str(y)[2:]
    
    # Respondents were interviewed every two year starting 1994
    # If the given year is not one of the survey years, find the subsequent survey year
    if not str_y in years:
        str_y = str(y+1)[2:]
        
    # The question enrolled in school since last interview was not asked until 1980
    in_school_since_last_int = None
    if y >= 1980:
        in_school_since_last_int = school_row['inschoolsincelastint'+str_y]
     
    # Get other relevant info: whether currently in school, age in that year, current grade in that year,
    # highest grade competed, reason for leaving school
    in_school_current = school_row['inschool'+str_y]
    age = age_row['ageatint'+str_y]
    current_grade = school_row['grade'+str_y]
    hi_grade = school_row['hi_grade_completed'+str_y]
    reason = school_row['reasonleftsch'+str_y]
    

        
    if y in [1978, 1979]:
        
        # There is no such question 'enrolled since last interview'
        
        if in_school_current == 0: # Not currently in school
            
            # Find last enrollment dates (should be valid responses in most cases)
            last_y = school_row['lastenroll_y'+str_y]            
            last_m = school_row['lastenroll_m'+str_y]

            if int(last_y) <= 0: # Missing last enrollment year
                counts['78-79, not currently in school, missing last enroll year'].append((person_id,y,m,age))
                return 0
                    
            else:
                if convert_year(last_y) < y:
                    return False
                elif convert_year(last_y) == y:
                    if last_m > 0:
                        if int(last_m) < m: 
                            return False
                        elif int(last_m) == m:
                            return True
                        else:
                            if hi_grade <= 12:
                                counts['78-79, not currently in school, last enroll month later than given month, hi grade <= 12, assumed True'].append((person_id,y,m,age))
                                return True
                            elif age <= 19:
                                counts['78-79, not currently in school, last enroll month later than given month, age under 19, assumed True'].append((person_id,y,m,age))
                                return True
                            else:
                                counts['78-79, not currently in school, last enroll month later than given month, undetermined'].append((person_id,y,m,age))
                                return 0
                    else:
                        counts['78-79, not currently in school, same year, missing last enroll month'].append((person_id,y,m,age))
                        return 0
                        
                elif convert_year(last_y) > y:
                    counts['78-79, not currently in school, last enrollment year later than given year'].append((person_id,y,m,age))
                    return 0
            
            
        elif in_school_current == 1: # Currently in school
            
            if age <= 19:
                counts['78-79, currently in school, age under 19, assumed True'].append((person_id,y,m,age))
                return True
            
            if current_grade < 0: # should be valid responses in most cases
                counts['78-79, currently in school, missing grade'].append((person_id,y,m,age))
                return 0
            elif current_grade <= 12:
                counts['78-79, currently in school, grade <= 12, assumed True'].append((person_id,y,m,age))
                return True
            else:
                counts['78-79, currently in school, undetermined'].append((person_id,y,m,age))
                return 0
    

    else:
        if in_school_since_last_int == 0:
        
            # There are last enrollment date variables before and in 1998
            # Monthly enrollment questions were asked only after 1980
            
            col = 'enrollsincelastint_'+calendar.month_name[m][:3].lower()+str_y0
            cols = find_header(school_df.columns, col) 
            data = [int(school_row[col]) for col in cols]
            
            if has_positive(data):
                counts['post80, not enrolled since last int, but has positive months'].append((person_id,y,m,age))
                return True
            else:
                return False
      
  
        elif in_school_since_last_int == 1: 

            # Find columns containing the given month and year                
            col = 'enrollsincelastint_'+calendar.month_name[m][:3].lower()+str_y0
            cols = find_header(school_df.columns, col) 
            
            # Find data in those columns
            data = [int(school_row[col]) for col in cols]
            
            # Decide whether R was in school in the given month
            if len(set(data)) == 1:
                
                if data[0] > 0:
                    return True
                else:
                    return False
                    
            elif len(set(data)) == 2:
                # Treat as in school as long as at least there is one valid in-school indicator
                
                if data[0] > 0 or data[1] > 0:
                    return True
    
                else:
                    return False
              
        
        elif in_school_since_last_int == -5: 
            
            # First, look for any monthly indicators
            # The status of the given month might have been asked in a later survey year
            col = 'enrollsincelastint_'+calendar.month_name[m][:3].lower()+str_y0
            cols = find_header(school_df.columns, col) 
            data = [int(school_row[col]) for col in cols if int(school_row[col])!=-5]
            
            assert len(data) < 2
            
            if len(data) == 1:
                if data[0] > 0:
                    return True
                elif data[0] in [0, -4]:
                    return False
        
            # Second, if the following (survey) year is in or before 1998, check out last enrollment dates
            next_yr = years[years.index(str_y)+1]
                
            if school_row['inschoolsincelastint'+next_yr] == -5:
                counts['post80, noninterview, noninterview in the following year as well'].append((person_id,y,m,age))
                return 0
            
            else:
                
                if years.index(next_yr) <= years.index('98'):
                    lastenroll_y = school_row['lastenroll_y'+next_yr]
                    lastenroll_m = school_row['lastenroll_m'+next_yr]
                    
                    if lastenroll_y > 0:
                        if convert_year(lastenroll_y) < y:
                            return False
                        
                        elif convert_year(lastenroll_y) == y:
                            if lastenroll_m > 0:
                                if lastenroll_m < m:
                                    return False
                                elif lastenroll_m == m: 
                                    return True
                                else:
                                    counts['post80, noninterview, same year, last enroll month later than given month'].append((person_id,y,m,age))
                                    return 0
                                    
                            else:
                                counts['post80, noninterview, same year, missing last enroll month'].append((person_id,y,m,age))
                                return 0
                                
                        elif convert_year(lastenroll_y) > y:
                            counts['post80, noninterview, last enroll year later than given year'].append((person_id,y,m,age))
                            return 0
                            
                    else:
                        counts['post80, noninterview, missing last enroll year'].append((person_id,y,m,age))
                        return 0
            
            # The left over cases
            counts['post80, noninterview, undetermined'].append((person_id,y,m,age))
            return 0
                    
                        
        else:
            counts['post80, other missing values'].append((person_id,y,m,age))
            return 0
        
```

##4. Determine quarterly status

```python
counts = defaultdict(set)
qrt_data = defaultdict(list)

for i, row in res_df_all.iterrows():
    
    if i%100 == 0:
        print i

    id = int(row['id'])
    qrt_data['id'].append(id)
        
    #13 weeks per quarter. 72 quarter per sequence
    for qrt_num in range(72):
        qrt_start_week = qrt_num*13
            
        #get the weekly array of employment status 
        wks_in_qrt = [row['week%s'%(qrt_start_week+j)] for j in range(13)]
                   
        # Determine quarterly status
        qrt_status = None
        
        s = list(set(wks_in_qrt))
        s.sort()
        
        if len(s) == 1:
            if s[0] == 0:
                counts['all 0s'].add(id)
                
            qrt_status = s[0]
                
        elif len(s) == 2:
            if s[0] == 0: # a missing and a non-missing
                if wks_in_qrt.count(s[1]) > wks_in_qrt.count(0):
                    qrt_status = s[1]
                else:
                    qrt_status = 0
                    counts['0 and a non-missing, 0 is more frequent'].add(id)
            else:
                ########################### Need to think about 1.5 and 3 ###########################
                if wks_in_qrt.count(s[0]) > wks_in_qrt.count(s[1]):
                    qrt_status = s[0]
                else:
                    qrt_status = s[1]
            
        elif len(s) == 3:
            if s[0] == 0: # a missing and 2 non-missings
                a, b, c = s
                
                assert a == 0, 'wrong assertion about order'
                
                ########################### Need to think about 1.5 and 3 ###########################
                
                if wks_in_qrt.count(b) > wks_in_qrt.count(c):
                    if wks_in_qrt.count(b) > wks_in_qrt.count(c) + wks_in_qrt.count(0):
                        qrt_status = b
                    else:
                        qrt_status = -1
                        counts['3 unique statuses with 0, most frequent status changed'].add(id)
                    
                elif wks_in_qrt.count(b) == wks_in_qrt.count(c):
                    qrt_status = -1
                    counts['3 unique statuses with 0, two non-missing statuses equally frequent'].add(id)
                else:
                    if wks_in_qrt.count(b) + wks_in_qrt.count(0) < wks_in_qrt.count(c):
                        qrt_status = c
                    else:
                        qrt_status = -1
                        counts['3 unique statuses with 0, most frequent status changed'].add(id)
                        
            else:
                # Define quarter status to be the most frequent status
                ########################### Need to think about 1.5 and 3 ###########################
                pairs = [(item,wks_in_qrt.count(item)) for item in s if item != 0]
                pairs = sorted(pairs,key=itemgetter(1),reverse=True)
                
                if pairs[0][1] > pairs[1][1]:
                    qrt_status = pairs[0][0]
                elif pairs[0][1] == pairs[1][1]:
                    qrt_status = -1
                    counts['3 unique statuses without 0, top two statuses equally frequent'].add(id)
                    
                
        elif len(s) > 3:
            
            if s[0] == 0:
                ########################### Need to think about 1.5 and 3 ###########################
                
                pairs = [(item,wks_in_qrt.count(item)) for item in s if item != 0]
                pairs = sorted(pairs,key=itemgetter(1),reverse=True)
                
                if pairs[0][1] > pairs[1][1] and pairs[1][1] > pairs[2][1]:
                    # The most frequest and second most frequent statuses are clearly determined
                    
                    if pairs[0][1] > pairs[1][1] + wks_in_qrt.count(0):
                        qrt_status = pairs[0][0]
                    else:
                        qrt_status = -1
                        counts['%s unique statuses with 0, most frequent status changed'%len(s)].add(id)
                        
                else:
                    qrt_status = -1
                    counts['%s unique statuses with 0, most frequent and second most frequent statuses hard to determine'%len(s)].add(id)
                    
            else:
                ########################### Need to think about 1.5 and 3 ###########################
                pairs = [(item,wks_in_qrt.count(item)) for item in s]
                pairs = sorted(pairs,key=itemgetter(1),reverse=True)
                
                if pairs[0][1] > pairs[1][1]: # The most freuqent status is clearly determined
                    qrt_status = pairs[0][0]
                else:
                    qrt_status = -1
                    counts['%s unique statuses without 0, most frequent status hard to determine'%len(s)].add(id)
                    
        
        assert qrt_status != None, 'quarter status still undetermined'
    
    
        # Save final quarterly status to dictionary
        qrt_data['qrt%s'%qrt_num].append(qrt_status)

            
```
